```python
cardio_base = pd.read_csv('/Users/yuliabezginova/PycharmProjects/00_files-for_NLP/data_turing/cardio_base.csv')
cardio_base.head()
```

```python
def desc_stat(data, column):
    # descriptive statistics
    print(data[column].describe())
    data.boxplot(column=column, figsize=(4, 10));
    plt.title('Descriptive Statistics')
    plt.show()
#     data[column].hist(bins=30, range=(0, 110), figsize=(15, 5));
```

```python
desc_stat(cardio_base, 'height')
```

```python
cardio_base = cardio_base.drop(cardio_base[(cardio_base['height'] >= 150) & (cardio_base['height'] <= 200)].index)
cardio_base.head()
```

```python
desc_stat(cardio_base, 'height')
```

```python
# using quantile() function to
# find the quantiles over the index axis
cardio_base.quantile([.1, .25, .5, .75, .99], axis = 0)
```
